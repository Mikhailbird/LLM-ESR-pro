```python
from collections import defaultdict

# 读取原始交互数据
user2items = defaultdict(list)
item2users = defaultdict(list)

with open("check_data/fashion/inter.txt") as f:
    for line in f:
        u, i = map(int, line.strip().split())
        user2items[u].append(i)
        item2users[i].append(u)

# 获取每个用户交互次数
user_lens = {u: len(items) for u, items in user2items.items()}
item_lens = {i: len(users) for i, users in item2users.items()}

# 尝试不同的 user_core / item_core 阈值
for uc in [1, 2, 5, 10, 15, 20]:
    for ic in [1, 2, 5, 10, 15, 20]:
        # 过滤用户和物品
        filtered_users = {u for u in user2items if len(user2items[u]) >= uc}
        filtered_items = {i for i in item2users if len(item2users[i]) >= ic}

        # 再次统计交互中满足过滤条件的 user/item 数量
        final_user_set = set()
        final_item_set = set()

        for u in filtered_users:
            for i in user2items[u]:
                if i in filtered_items:
                    final_user_set.add(u)
                    final_item_set.add(i)

        print(f"user_core={uc}, item_core={ic} -> users: {len(final_user_set)}, items: {len(final_item_set)}")
```
